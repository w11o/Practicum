```python
#5.2
def f(*args: int | float | tuple | list) -> int | float | tuple | list:
    res = 1
    if type(args[0]) == int or type(args[0]) == float:
        for i in args:
            res *= i
        return res

    if type(args[0]) == tuple or type(args[0]) == list:
        for i in args[0]:
            res *= i
        return res

print(f(3))
print(f(1, 2))
print(f(1, 2, 3))
```

```python
#5.5
def f(num1:int, num2:int, operation:int) -> int:
    if operation == 1:
        return int(num1 + num2)

    if operation == 2:
        return int(num1 - num2)

    if operation == 3:
        return int(num1 * num2)

    if operation == 4:
        return int(num1 / num2)

print('введите число 1')
num1 = int(input())

print('введите число 2')
num2 = int(input())

print('выберите операцию:'
      '\n1.сложение'
      '\n2.вычитание'
      '\n3.умножение'
      '\n4.деление')

operation = int(input())

print(f(num1, num2, operation))
```

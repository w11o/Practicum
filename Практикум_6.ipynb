```python
def gen_rectangle(start):
    fig, ax = plt.subplots()
    
    for _ in range(20):
        poly = Polygon(start, facecolor='none', edgecolor='blue')
        ax.add_patch(poly)
        start = [(x + 0.1, y) for (x, y) in start]
        print(start)
        

    ax.set_xlim([0, 2])
    ax.set_ylim([0, 2])
    
gen_rectangle([(0.1, 0.01), (0.1, 0.2), (0.2, 0.2), (0.2, 0.01)])
```

```python
def tr_translate(start, up = None, down = None):
    fig, ax = plt.subplots()
    start = [(x, y + up) for (x, y) in start]

    for _ in range(20):
        poly = Polygon(start, facecolor='none', edgecolor='blue')
        ax.add_patch(poly)
        start = [(x + 1.1, y) for (x, y) in start]
        

    ax.set_xlim([0, 10])
    ax.set_ylim([0, 10])
tr_translate([(0.1, 0.6), (0.4, 0.1), (0.8, 0.1), (1.1, 0.6), (0.8, 1.1), (0.4, 1.1)], up = 0.5) 
```

```python
def tr_rotate(start, up = None, down = None):
    fig, ax = plt.subplots()

    if up:
        for _ in range(20):
            poly = Polygon(start, facecolor='none', edgecolor='blue')
            ax.add_patch(poly)
            start = [(x + 1.1, y + up) for (x, y) in start]

    if down:
        for _ in range(20):
            poly = Polygon(start, facecolor='none', edgecolor='blue')
            ax.add_patch(poly)
            start = [(x + 1.1, y - down) for (x, y) in start]

        

    ax.set_xlim([0, 10])
    ax.set_ylim([0, 10])
tr_rotate([(0.1, 0.6), (0.4, 0.1), (0.8, 0.1), (1.1, 0.6), (0.8, 1.1), (0.4, 1.1)], up = 0.5) 
```

```python
def tr_three(start, up = None, down = None):
    fig, ax = plt.subplots()
    begin = start
    for _ in range(3):
        for _ in range(20):
            poly = Polygon(start, facecolor='none', edgecolor='blue')
            ax.add_patch(poly)
            start = [(x + 0.1, y + up) for (x, y) in start]
        begin = [(x, y + 0.5) for (x, y) in begin]
        start = begin


        

    ax.set_xlim([0, 2])
    ax.set_ylim([0, 2])
tr_three([(0.1, 0.01), (0.1, 0.2), (0.2, 0.2), (0.2, 0.01)], up = 0.09) 
```

```python
def gen_rectangle(start):
    fig, ax = plt.subplots()
    
    for _ in range(20):
        poly = Polygon(start, facecolor='none', edgecolor='blue')
        ax.add_patch(poly)
        start = [(x + 0.1, y) for (x, y) in start]
        print(start)
        

    ax.set_xlim([0, 2])
    ax.set_ylim([0, 2])
    
gen_rectangle([(0.1, 0.01), (0.1, 0.2), (0.2, 0.2), (0.2, 0.01)])
```
